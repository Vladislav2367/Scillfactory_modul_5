```python
# Первоначальное окно

def first_window():
    print("")
    print("")
    print("                                                 // TIC-TAC-TOE Game //     ")
    print("                                                     крестики-нолики  ")
    print("")
    print("                                                      Порядок ввода: ")
    print("")
    print("                                                     x - номер строки  ")
    print("                                                     y - номер столбца ")
first_window()


# Массив пространства игры

array_game = [[" ", " ", " "] for i in range(3)]


# Рисуем клеточки

def cells():
    print()
    print("                                                    Y\X| 0 | 1 | 2 | ")
    print("                                                     --------------- ")
    for i, row in enumerate(array_game):
    
        row_str = f"                                                    {i}  | {' | '.join(row)} | "
        print(row_str)
        print("                                                     --------------- ")
    
cells()


# Координаты ходов

def coordin():
    while True:
        x = input("                  Введите координату клеточки по Х (от 0 до 2): ",)
        y = input("                  Введите координату клеточки по Y (от 0 до 2): ",)
                      
        if not(x.isdigit()) or not(y.isdigit()):
            print("")
            print("                  Внимание! Необходимо число от 0 до 2")
            continue
        
        x = int(x)
        y = int(y)
        
        if 0 > x or x > 2 or  0 > y or  y > 2 :
            print("")
            print("                 //Внимание! Необходимо число от 0 до 2//")
            continue
        
        if array_game[x][y] != " ":
            print("")
            print("                 // Клетка занята!!! // ")
            continue
        
        return x, y
    
coordin()


# Определение победителя

def check_hero():
    sim2 = 0
    sim3 = 0
    sim4 = 0
    sim5 = 0
    for i in range (3):
        sim0 = 0
        sim1 = 0
        for j in range (3):

            if array_game[i][j] == "X":   #
                sim0 +=1                  #
            if array_game[i][j] == "0":   #
                sim1 +=1                  ## построчная проверка наличия выигравших
            if sim0 == 3:                 #
                print("Выиграл X!!!")     #
                return True               #
            if sim1 ==3:                  #
                print("Выиграл 0!!!")     #
                return True

        # проверка наличия выигравших по восходящей диагонали

            if array_game[i][j] == "X" and i == j:
                sim2 +=1
            if array_game[i][j] == "0" and i == j:
                sim3 +=1
            if sim2 == 3:
                print("Выиграл X!!!")
                return True
            if sim3 ==3:
                print("Выиграл 0!!!")
                return True

        # проверка наличия выигравших по нисходящей диагонали
        
            if array_game[i][j] == "X" and i == 2 - j:
                sim4 +=1
            if array_game[i][j] == "0" and i == 2 - j:
                sim5 +=1
            if sim4 == 3:
                print("Выиграл X!!!")
                return True
            if sim5 ==3:
                print("Выиграл 0!!!")
                return True

    return False
            
check_hero()


# Алгоритм ходов

first_window()
array_game = [[" ", " ", " "] for i in range(3)]

k = 0
while True:
    k += 1
    cells()
    if k % 2 == 0:
        print(" Ходит нолик!")
    else:
        print(" Ходит крестик!")
    
    x,y = coordin()
    
    if k % 2 == 0:
        array_game[x][y] = "0"
    else:
        array_game[x][y] = "X"

    if check_hero():
        break

    if k == 9:
        print(" Ничья!")
        break       
```
